```python
# # try to handle page 1

import requests 
from bs4 import BeautifulSoup

r = requests.get("https://www.price.com.hk/search.php?g=A&q=dyson&page=1")
soup = BeautifulSoup(r.text)
```

```python
# product name 

product_names = []
for item_inner in soup.findAll('div', {'class': 'item-inner'}):
    line_line_01 = item_inner.find('div', {'class': 'line line-01'})
    tag_a = line_line_01.find('a')
    product_names.append(tag_a.text.strip())
product_names

#product price , only 行貨


A_product_price = []

for price_range in soup.findAll('div', {'class': 'line line-01 price-range'}):
    tag_b = price_range.find('a')
    tag_c = tag_b.find('span', {'class': 'product-prop'})

    if tag_c:
        tag_d = tag_c.find("img")
        Title = tag_d.get("title") if tag_d else None

        if Title == '行貨':
            product_price = tag_b.text.strip()
            A_product_price.append(('行貨:', product_price))


A_product_price 


#create a dict to store product  name and price 

catalogues = {}

for i in range(len(product_names)):
    catalogues[product_names[i]] = A_product_price[i]

catalogues
    

```
